### **Floor division operator** (```//```)

```x//y``` = floor(x/y)

####**The Floor function**


- Floor function gives the biggest integer less than the value.
- On a number line, it gives the closest integer to he left of the value.

- **floor(-0.67)*** = -1

- **floor(2.3)** = 2

### **Modulus Operator** (```%```)




```x % y``` -> remainder of `x / y`
- If `x` is **'+ve'** -> remainder of `x / y`
- If `x` is **'-ve'** -> `y - (x % y)`

````python
10 % 3 #1
````

````python
-10 % 3 #2
````
