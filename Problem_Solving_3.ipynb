```python
class Solution:
    def solve(self, A, B):
        for i in range(len(A)):
            if A[i]>=B:
                ceil = A[i]
                floor = A[i-1]
                break
            elif A[i]==B:
                ceil = A[i]
                floor= A[i]
                break

        return ([ceil,floor])
```
